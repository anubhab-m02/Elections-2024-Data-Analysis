# Data Scraping using BeautifulSoup

## Scraping the Party Wise Results

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = 'https://results.eci.gov.in/PcResultGenJune2024/index.htm'

response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')

table = soup.find('table')

data = [["Party", "Seats Won", "Leading", "Total"]]
for row in table.find_all('tr'):
    cols = row.find_all('td')
    cols = [ele.text.strip() for ele in cols]
    data.append([ele for ele in cols if ele])
df = pd.DataFrame(data)

df.to_csv('Lok_Sabha_Election_Results_2024.csv', index=False)
```

## Scraping the Constituency Wise Results

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = 'https://results.eci.gov.in/PcResultGenJune2024/index.htm'

response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')

my_dict = {}
options_states_and_ut = soup.find_all('option')
options_states_and_ut.remove(options_states_and_ut[0])
data=[["State", "Constituency","Leader", "Party", "Margin", "Total Votes"]]
for option in options_states_and_ut:
    try:      
        my_dict[option.text] = option['value']
        url_cons = "https://results.eci.gov.in/PcResultGenJune2024/partywiseresult-"+option['value']+".htm"
        response_cons = requests.get(url_cons)
        soup_cons = BeautifulSoup(response_cons.content, 'html.parser')
        cons_options = soup_cons.find_all('option')
        cons_options.remove(cons_options[0])
        for cons in cons_options:
            try:    
                url3 = "https://results.eci.gov.in/PcResultGenJune2024/Constituencywise"+cons['value']+".htm"
                
                response3 = requests.get(url3)
                soup3 = BeautifulSoup(response3.content, 'html.parser')
                leading = soup3.find_all('tr')[1].find_all('td')[1].text
                
                leading_party = soup3.find_all('tr')[1].find_all('td')[2].text
                
                margin = int(soup3.find_all('tr')[1].find_all('td')[5].text) - int(soup3.find_all('tr')[2].find_all('td')[5].text)
                
                total_votes = soup3.find_all('th')[-2].text
                
                data.append([option.text,cons.text,leading,leading_party,margin, total_votes])
                print(option.text+"\t"+cons.text+"\t"+leading+"\t"+leading_party+"\t"+str(margin)+"\t"+str(total_votes))
                print("\n")
                
            except:
                pass
    except:
        pass
    
df = pd.DataFrame(data)
df.to_csv('Constituency_Wise.csv', index=False)
```
